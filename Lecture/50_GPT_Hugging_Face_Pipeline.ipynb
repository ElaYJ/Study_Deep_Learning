```python
from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis')
```

```python
classifier.model
```

zero-shot Model

```python
classifier = pipeline("zero-shot-classification")
```

```python
classifier(
    "This is a course about the Transformers library",
    candidate_labels = ['education','politics','business']
)
```

```python
classifier(
    "This is a course about the Transformers library",
    candidate_labels = ['libgame','game','please','library','computer game','computer','programming']
)
```

생성 모델

```python
generator = pipeline("text-generation")
```

```python
# 다음에 가장 어울리는 문장 생성

generator("In this course, we will teach you how to")
```

```python
generator("In this course, we will teach you how to", num_return_sequences=2, max_length=30)
```

```python
list_ = ["In this course, we will teach you how to", 'This is a course about the Transformers library']

for sentence in list_:
    print(generator(sentence, num_return_sequences=1, max_length=50))
```

Default 모델 말고 특정 모델을 사용해본다.

Hugging Face site에 있는 다양한 모델을 사용해볼 수 있다.

```python
generator = pipeline("text-generation", model="huggingtweets/_nisagiss-dril_gpt2-drilbot_neo")
```

```python
generator("My dream is", num_return_sequences=5)
```

```python
for sentence in list_:
    print(generator(sentence, num_return_sequences=1, max_length=50))
```

- QnA

- \<Query text> 나는 집에 들어갔다. 그런데 배가 고파서 햄버거를 먹었다.
    Q: 누가 햄버거를 먹었나요?

    A: 나 --> query text에 있는 답을 가져올 수 있다.

    => 주로 모델링된 자연어 처리이다.

- \<Query text> 우리 가족은 집에서 잠을 자고 있었다. 나는 게임중이었는데, 너무 배가 고팠다.

    Q: 내가 다음에 할 행동은?

    A: 먹는다. --> query text에 없는 답을 찾아야 한다.

    => 아직은 한계가 존재

```python
qna = pipeline("question-answering")
```

```python
qna(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn.",
)
```
