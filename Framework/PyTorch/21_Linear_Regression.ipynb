```python
df_target = df.iloc[:, -1:]
df_target
```

```python
x = torch.tensor(df.values)
y = torch.tensor(df_target.values).view(-1, 1)
x.shape, y.shape
```

```python
XT = torch.transpose(x, 0, 1)
XT.shape
```

```python
w = torch.matmul(torch.matmul(torch.inverse(torch.matmul(XT, x)), XT), y)
w
```

```python
w = torch.mm(torch.mm(torch.linalg.inv(torch.mm(XT, x)), XT), y)
w
```

#### assign 대신에 data에 접근해서 값을 수정

tensor.data = 다른데이터

```python
w
```

```python
w.data
```

```python
from IPython.display import display, clear_output
```

```python
learning_rate = 3e-7

for epoch in range(100):
    z = x.matmul(w) + b
    loss = torch.mean((y - z)**2)
    loss.backward()
    w.data -= learning_rate * w.grad
    b.data -= learning_rate * b.grad

    # print("{:3} - loss : {}".format(epoch, loss.item()), end="\r")
    display("{:3} - loss : {}".format(epoch, loss.item()))
    clear_output(wait=True)

    # 이전 값을 누적하지 않도록 초기화 필요
    w.grad.zero_()
    b.grad.zero_()
```

```python
learning_rate = 0.000003

for epoch in range(100):
    z = (x.matmul(w) + b)
    loss = torch.mean((z - y)**2)

    grads = torch.autograd.grad(loss, [w, b])
    w.data -= grads[0] * learning_rate
    b.data -= grads[1] * learning_rate

    # print("{:3} - loss : {}".format(epoch, loss.item()), end="\r")
    display("{:3} - loss : {}".format(epoch, loss.item()))
    clear_output(wait=True)
```

#### optimizer 사용하기

```python
opt = torch.optim.SGD([w, b], lr=learning_rate)
# [w, b] : 미분값을 추적할 파라미터들
```

```python
for epoch in range(100):
    z = (x.matmul(w) + b)
    loss = torch.mean((z - y)**2)

    loss.backward()
    opt.step()
    # print("{:3} - loss : {}".format(epoch, loss.item()), end="\r")
    display("{:3} - loss : {}".format(epoch, loss.item()))
    clear_output(wait=True)

    opt.zero_grad()
```

```python
y_pred = x.matmul(w) + b
print("예측한 집값 :", y_pred[19].item(), "실제 집값 :", df_target.values[19])
```
