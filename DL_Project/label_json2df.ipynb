```python
from datetime import datetime
from tqdm.notebook import tqdm

def convert_json2df(src_list, df_data):
    for jname in tqdm(src_list, desc='Processing'):
        with open(os.path.join(json_data_dir, jname), 'r') as f:
            data = json.load(f)

        f_name = jname.split('.')[0]
        df_data['fname'].append(f_name)
        val = data['images']['image_id']
        df_data['image_id'].append(val if val else None)
        val = data['images']['farm_id']
        df_data['farm_id'].append(val if val else None)
        val = data['images']['crops_id']
        df_data['crops_id'].append(val if val else None)
        val = datetime.strptime(data['images']['date_captured'], "%Y-%m-%d %H:%M:%S")
        df_data['c_date'].append(val.date() if val else None)
        df_data['c_time'].append(val.time() if val else None)
        val = data['images']['crops'] +" "+ data['images']['kind_type']
        df_data['crops_type'].append(val if val else None)
        val = data['images']['growth_stage']
        df_data['growth_stage'].append(val if val else None)
        val = data['envrionments'][0]
        df_data['env1'].append(val if val else None)
        val = data['envrionments'][1]
        df_data['env2'].append(val if val else None)
        val = data['growth_index']
        df_data['growth_info'].append(val if val else None)
        val = data['etc_infor']
        df_data['etc_info'].append(val if val else None)
        val = int(data['images']['width'])
        df_data['img_width'].append(val if val else None)
        val = int(data['images']['height'])
        df_data['img_height'].append(val if val else None)
        anno = data['annotations']
        df_data['bbox_num'].append(len(anno))
        df_data['annotations'].append(anno)
        cate = data['categories']
        df_data['categories'].append(cate)
```

```python
df_data = dict(
    fname=[], image_id=[], farm_id=[], crops_id=[],
    crops_type=[], growth_stage=[], c_date=[], c_time=[],
    env1=[], env2=[], growth_info=[], etc_info=[],
    img_width=[], img_height=[], bbox_num=[], annotations=[], categories=[]
)

convert_json2df(json_list, df_data)
```

```python
import pandas as pd

df = pd.DataFrame(df_data)
df
```

```python
df.to_csv("/content/drive/MyDrive/DLP/[DS]_Deep_Learning_Project/train_meta_data.csv", index=False)
```

# info data

## train

```python
import pandas as pd

tex_df = pd.read_csv("/content/drive/MyDrive/DLP/dataset/train_ex_data.csv")
tex_df
```

```python
tfname_list = tex_df['fname'].values.tolist()
len(tfname_list)
```

```python
import os
from tqdm.notebook import tqdm

json_dir = "/content/drive/MyDrive/DLP/dataset/label_data_json/decomp_train/"

for fname in tqdm(tfname_list):
    if not os.path.isfile(json_dir + fname + ".json"):
        raise ValueError("Json file not found:", fname)
```

```python
import json
from datetime import datetime

def get_info_from_json(json_fnames):
    base_data = dict(
        fname=[], crops_id=[], crops_type=[], growth_stage=[], c_date=[], c_time=[],
        img_width=[], img_height=[], bbox_num=[], class_num=[], class_name=[]
    )
    env1_data=[]; env2_data=[]; growth_info=[];

    for f_name in tqdm(json_fnames, desc='Processing'):
        with open(os.path.join(json_dir, f_name+".json"), 'r') as f:
            data = json.load(f)

        base_data['fname'].append(f_name)
        val = data['images']['crops_id']
        base_data['crops_id'].append(val if val else None)
        val = data['images']['crops'] +" "+ data['images']['kind_type']
        base_data['crops_type'].append(val if val else None)
        val = data['images']['growth_stage']
        base_data['growth_stage'].append(val if val else None)
        val = datetime.strptime(data['images']['date_captured'], "%Y-%m-%d %H:%M:%S")
        base_data['c_date'].append(val.date() if val else None)
        base_data['c_time'].append(val.time() if val else None)
        val = int(data['images']['width'])
        base_data['img_width'].append(val if val else None)
        val = int(data['images']['height'])
        base_data['img_height'].append(val if val else None)
        anno = data['annotations']
        base_data['bbox_num'].append(len(anno))
        cate = data['categories']
        class_name = set()
        for c in cate:
            class_name.add(c['name'])
        base_data['class_num'].append(len(class_name))
        base_data['class_name'].append(class_name)

        e1 = dict(fname=f_name)
        e1.update(data['envrionments'][0])
        env1_data.append(e1)

        e2 = dict(fname=f_name)
        e2.update(data['envrionments'][1])
        env2_data.append(e2)

        gi = dict(fname=f_name)
        gi.update(data['growth_index'])
        gi.update(data['etc_infor'])
        growth_info.append(gi)

    return base_data, env1_data, env2_data, growth_info
```

```python
base_data, env1_data, env2_data, growth_info = get_info_from_json(tfname_list)
```

```python
base_df = pd.DataFrame(base_data)
base_df
```

```python
env2_df = pd.DataFrame(env2_data)
env2_df = env2_df.drop(columns=['farm_id','id','tl_value','ei_value','pl_value'])
```

```python
numeric_columns = ['ti_value', 'hi_value', 'ci_value', 'ir_value', 'sr_value', 'cl_value', 'el_value', 'hl_value', 'pi_value']
env2_df[numeric_columns] = env2_df[numeric_columns].apply(pd.to_numeric)
env2_df
```

```python
growth_info_df = pd.DataFrame(growth_info)
growth_info_df
```

```python
growth_info_df['fr1_cnt'] = growth_info_df['fr1_cnt'].map(lambda x: 0 if x == None else x)
growth_info_df['fr2_cnt'] = growth_info_df['fr2_cnt'].map(lambda x: 0 if x == None else x)
growth_info_df['fr3_cnt'] = growth_info_df['fr3_cnt'].map(lambda x: 0 if x == None else x)
growth_info_df['fr_weight'] = growth_info_df['fr_weight'].map(lambda x: 0 if x == None else x)
```

```python
numeric_columns = [
    'stem_length', 'leaf_cnt', 'leaf_width', 'leaf_length', 'stem_thick',
    'fr1_cnt', 'fr2_cnt', 'fr3_cnt', 'fr_weight'
]
growth_info_df[numeric_columns] = growth_info_df[numeric_columns].apply(pd.to_numeric)
growth_info_df['fr_cnt'] = growth_info_df['fr1_cnt'] + growth_info_df['fr2_cnt'] + growth_info_df['fr3_cnt']

growth_info_df.drop(
    columns=[
        'corps_id','bloom1_date','bloom2_date','bloom3_date',
        'fr1_cnt','fr2_cnt','fr3_cnt','create_date','inform'
    ],
    inplace=True
)
growth_info_df
```

```python
df = base_df.merge(growth_info_df, how='left', on='fname')
df = df.drop(columns=['img_width','img_height','measured_date'])
df
```

```python
weights_df = df.query("fr_weight>0 & fr_cnt>0")
weights_df
```

```python
weights_df.to_csv("/content/drive/MyDrive/DLP/dataset/train_weights.csv", index=False)
```

```python
df = base_df.merge(growth_info_df, how='left', on='fname')
df = df.merge(env2_df, how='left', on='fname')
df = df.drop(columns=['measured_date','receive_date'])
df.info()
```

```python
df.to_csv("/content/drive/MyDrive/DLP/dataset/train_growth_env_info.csv", index=False)
```
